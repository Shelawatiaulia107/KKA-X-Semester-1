#Aplikasi Menghitung Luas Persegi Panjang

```python
panjang = int(input("Masukkan panjang persegi panjang (cm): "))

lebar = int(input("Masukkan lebar persegi panjang (cm): "))

luas = panjang * lebar

print (f"Luas persegi panjang yaitu {luas} cm^2.")
```

```python
pi = 3.14
Jari_jari = int(input("Masukkan_jari_jari_tabung (cm) : "))
tinggi = int(input("Masukkan_tinggi_tabung (cm): "))

volume = pi * Jari_jari * Jari_jari * tinggi
print(f"Volume_tabung_yaitu_{volume}_cm^3.")
```
