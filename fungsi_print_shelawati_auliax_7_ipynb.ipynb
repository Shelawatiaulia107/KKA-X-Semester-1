#Parameter Fungsi print

```python
Tanggal =15
Bulan ="Januari"
Tahun =2010

#15-Januari-2010
print(Tanggal)
print(Bulan)
print(Tahun)
```

```python
#15-Januari-2010
print(Tanggal, Bulan, Tahun, sep="-")
```

```python
#15-Januari-2010
print(Tanggal, end="-")
print(Bulan, end="-")
print(Tahun)
```
